```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
X = data.data
y = data.target
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=17)
```

```python
tree.fit(X, y)
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pydot
from IPython.core.display import Image, display
```

```python
export_graphviz(
    tree,
    feature_names=data.feature_names,
    out_file='iris_tree.dot',
    filled=True)
# для этого понадобится библиотека pydot (pip install pydot)
```

```python
graph, = pydot.graph_from_dot_file('iris_tree.dot')
graph.write_png('iris_tree.png')

display(Image('iris_tree.png', unconfined=True))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
model = RandomForestClassifier(max_features=2)
model.fit(X, y)
print(classification_report(y_true=y, y_pred=model.predict(X)))
```
